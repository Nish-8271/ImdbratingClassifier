```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### Sentences
sent=["the glass of milk","the glass of juice","the cup of tea","i am a good boy","i am a good developer",
      "understand the meaning of words","your videos are good"]
```

```python
sent
```

```python
### Vocab_size
vocab_size=500
```

```python
one_hot_repr=[one_hot(words,vocab_size) for words in sent]
one_hot_repr
```

```python
### WOrd Embedding Representation 
from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing import pad_sequence
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## Maximu  sentences length
sent_length=8
```

### Padding 
> It can be Pre or Post
> It is must as in the input layer of RNN the vector length at each timesequence must be equal  

```python
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
## Feature Representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_shape=(sent_length,)))
model.compile('adam','binary_crossentropy')
```
